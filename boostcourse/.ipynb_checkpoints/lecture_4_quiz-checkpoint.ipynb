```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/NHIS_OPEN_GJ_2017.CSV.zip",
                 encoding="cp949") # 기본적인 encoding = 'UTF-8'

df.shape
df['시력(좌)'].describe()
```

### 1번문제

```python
age_code = {1: '0~4세',
 2: '5~9세',
 3: '10~14세',
 4: '15~19세',
 5: '20~24세',
 6: '25~29세',
 7: '30~34세',
 8: '35~39세',
 9: '40~44세',
 10: '45~49세',
 11: '50~54세',
 12: '55~59세',
 13: '60~64세',
 14: '65~69세',
 15: '70~74세',
 16: '75~79세',
 17: '80~84세',
 18: '85세+'}

df = df.replace({'연령대코드(5세단위)':age_code})
df = df.rename(columns = {'연령대코드(5세단위)':'연령대'})
df.groupby(['연령대'])['허리둘레'].describe()
```

### 2번문제

```python
df_2 = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/NHIS_OPEN_GJ_2017.CSV.zip",
                 encoding="cp949") # 기본적인 encoding = 'UTF-8'
```

```python
df_2.head()
columns = ['음주여부', '흡연상태', '연령대코드(5세단위)', '성별코드']
df_small = df_2[columns]
df_corr = df_small.corr()
```

```python
plt.figure(figsize = (10, 7))
mask = np.triu(np.ones_like(df_corr, dtype = np.bool))
sns.heatmap(data = df_corr, annot = True, fmt = '.2f', mask = mask, cmap = 'Blues')
```

### 3번문제

```python
smoke = {1 : "흡연안함", 2: "끊음", 3: "흡연중"}
drink = {0: "안마심", 1: "마심"}

df = df.replace({'음주여부':drink})
df = df.replace({'흡연상태':smoke})
df = df.rename(columns = {'흡연상태':'흡연'})
df = df.rename(columns = {'음주여부':'음주'})
pd.crosstab(df.음주, df.흡연)
```

```python
sns.countplot(data = df, x = '흡연', hue = '음주')
```

### 4번문제

```python
df_over_119 = df[df['체중(5Kg 단위)'] > 119]
sns.lmplot(data = df_over_119, x = '총콜레스테롤', y = '감마지티피', hue = '음주', col = '흡연')
```

### 5번문제

```python
df_seeable = df[(df['시력(좌)'] < 9.9) & (df['시력(우)'] < 9.9)]
df_seeable = df_seeable.sort_values('연령대')

fig, a = plt.subplots(1, 2)
plt.subplots_adjust(left = 5, right = 7)

sns.barplot(data = df_seeable, x = '시력(좌)', y = '연령대', hue = '성별코드',ci = None, ax = a[0])
sns.barplot(data = df_seeable, x = '시력(우)', y = '연령대', hue = '성별코드',ci = None, ax = a[1])

```
