## Q2.

```python
# "동물병원"이 "상권업종소분류명"에 존재하는 지 확인한다.

df["상권업종소분류명"].unique()
```

```python
#  "df_vet" 변수에 "상권업종소분류명"이 "동물병원"인 데이터를 담는다.
# 전체 데이터에 영향을 주지 않도록 .copy()를 사용한다.
# vet(= veterinarian) : 수의사 / 동물병원

df_vet= df[df["상권업종소분류명"] == "동물병원"].copy()
df_vet.head()
```

### Q2. RESULT

```python
# value_counts()를 사용하여 전국 시도별 동물병원 수를 구한다.

df_vet_counts = df_vet["시도명"].value_counts()
df_vet_counts
```

## Q3

```python
# "피부"나 "성형"이 "상권업종소분류명"에 존재하는 지 확인한다.

df["상권업종소분류명"].unique()
```

```python
# 전체 데이터에서 "서울"에 위치하고 "상권업종중분류명"이 "병원"인 자료만 추출한다.
# df_seoul_hospital 변수에 담는다

df_seoul_hospital = df[(df["시도명"] == "서울특별시") & (df["상권업종중분류명"] == "병원") ].copy()
df_seoul_hospital.head()
```

```python
# df_seoul_hospital에서 "상권업종소분류명"이 "피부과", "정형/성형외과"인 자료를 추출한다.
# df_seoul_beauty 변수에 담는다.

df_seoul_beauty = df_seoul_hospital[(df_seoul_hospital["상권업종소분류명"] == "피부과") |
                           (df_seoul_hospital["상권업종소분류명"] == "정형/성형외과")].copy()
df_seoul_beauty.head()
```

```python
# "시군구명" 별 피부과·성형외과 수 / 전체병원 수 의 비율을 구한다.

n = df_seoul_beauty["시군구명"].value_counts()/df_seoul_hospital["시군구명"].value_counts()
n
```

### Q3. RESULT

```python
# 비률이 높은 순서대로 정렬하고, 소숫점 두 번째 자리까지 출력한다.

df_seoul_beauty_count = n.sort_values(ascending = False).round(2)
df_seoul_beauty_count
```

## Q4.

```python
# folium 을 불러온다.

import folium
```

```python
# "상권업종소분류명"이 "노인/치매병원"인 데이터를 추출하여
# df_nursing_house 변수에 담는다.

df_nursing_house = df[df["상권업종소분류명"] == "노인/치매병원"].copy()
df_nursing_house.shape
```

### Q4. RESULT

```python
# df_nursing_house의 위도와 경도의 평균값을 통해서 지도상에 대한민국이 표시되도록 한다.
# 반복문을 사용하여 Marker를 찍는다


map = folium.Map(location = [df_nursing_house["위도"].mean(), df_nursing_house["경도"].mean()], zoom_start = 6)

for n in df_nursing_house.index :
    name = df_nursing_house.loc[n, "상호명"]
    address = df_nursing_house.loc[n, "도로명주소"]
    popup = f"{name}-{address}"
    location = [df_nursing_house.loc[n, "위도"], df_nursing_house.loc[n, "경도"]]
    folium.Marker(location = location, popup = popup).add_to(map)
map
```
